```python
# Import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Create an URL object
url = 'https://covid19.karnataka.gov.in/new-page/Fever%20Clinics/en'
# Create object page
page = requests.get(url)
print(page)
# response 200 means the server allows us to collect data from their website
```

```python
# parser-lxml = Change html to Python friendly format
# Obtain page's information
soup = BeautifulSoup(page.text, 'lxml')
soup
```

```python
# Obtain information from tag <table>
table1 = soup.find('table')
table1
```

```python
# to get the headers:
# Obtain every title of columns with tag <th>
headers = []
for i in table1.find_all('td'):
 title = i.text
 headers.append(title)
    
headers = [i.strip('\n') for i in headers[:7]]
headers
```

```python
content = []
for i in table1.find_all('td'):
    info = i.text
    content.append(info)
    
content = [i.strip('\n') for i in content[8:]]
content
# we get content of all the fever clinics from here
```
